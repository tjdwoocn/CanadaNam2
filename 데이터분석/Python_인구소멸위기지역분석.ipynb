```python
population = pd.read_excel('C:/Users/tjdwo/pydata/05. population_raw_data.xlsx', header=1)
population
```

### tip1. df.fillna(0): 결측치 값을 0으로 채우기, df.fillna('missing'): 결측치 값을 'missing' 문자열로 채우기
### tip2. df.fillna(method='pad' or 'fill'), df.fillna(method='bfill' or 'backfill') : 결측치 값을 앞/뒤 방향으로 채우기
### tip3. inplace = true: 또 다른 객체를 반환하지 않고 기존 객체를 수정

```python
population.fillna(method='pad', inplace=True)
population.head()
```

```python
population.rename(columns={'행정구역(동읍면)별(1)': '광역시도', '행정구역(동읍면)별(2)':'시도', '계':'인구수'}, inplace=True)
population.head()
```

```python
population = population[population['시도'] != '소계']
population.head()
```

```python
population.is_copy=False

population.rename(columns={'항목':'구분'}, inplace=True)

population.loc[population['구분']=='총인구수 (명)', '구분'] = '합계'
population.loc[population['구분']=='남자인구수 (명)', '구분'] = '남자'
population.loc[population['구분']=='여자인구수 (명)', '구분'] = '여자'

population.head()
```

```python
# make rule
population['20-39세'] = population['20 - 24세'] + population['25 - 29세'] + \
                        population['30 - 34세'] + population['35 - 39세']

population['65세 이상'] = population['65 - 69세'] + population['70 - 74세'] + \
                          population['75 - 79세'] + population['80 - 84세'] + \
                          population['85 - 89세'] + population['90 - 94세'] + \
                          population['95 - 99세'] + population['100+']

population.head()
```

### tip1. df.is_copy = False :  데이터 처리 과정에서 발생하는 warning 이 뜨지 않도록 설정
### tip2. df.head(N): 상위 N개 로우 리턴, df.tail(N): 하위 N개 로우 리턴

```python
pivot = pd.pivot_table(population, index= ['광역시도', '시도'], 
                                  columns = ['구분'],
                                  values = ['인구수', '20-39세', '65세 이상'])
pivot

# tip. https://pandas.pydata.org/pandas-docs/stable/generated/pandas.pivot_table.html
```

### 비즈니스 로직에 따라 소멸비율과 소멸위기지역 컬럼 생성

```python
pivot['소멸비율'] = pivot['20-39세', '여자'] / (pivot['65세 이상', '합계'] / 2)
pivot['소멸위기지역']= pivot['소멸비율'] < 1.0

pivot
```

### 소멸위기지역을 기준으로 index 재생성

```python
pivot[pivot['소멸위기지역'] == True].index.get_level_values(1)
pivot.reset_index(inplace=True)
pivot.head()
```

### 컬럼값 재설정

```python
tmp_columns = [pivot.columns.get_level_values(0)[n] + \
               pivot.columns.get_level_values(1)[n] 
               for n in range(0,len(pivot.columns.get_level_values(0)))]

pivot.columns = tmp_columns
pivot
```

```python
pivot['시도'].unique()

si_name = [None] * len(pivot)

tmp_gu_dict = {
              '수원':['장안구', '권선구', '팔달구','영통구'],
              '성남':['수정구', '중원구', '분당구'],
              '안양':['만안구', '동안구'],
              '안산':['상록구', '단원구'],
              '고양':['덕양구', '일산동구', '일산서구'],
              '용인':['처인구', '기흥구', '수지구'],
              '청주':['상당구', '서원구', '흥덕구', '청원구'],
              '천안':['동남구', '서북구'],
              '전주':['완산구', '덕진구'],
              '포항':['남구', '북구'],
              '창원':['의창구', '성산구', '진해구', '마산합포구', '마산회원구'],
              '부천':['오정구', '원미구', '소사구']
              }

tmp_gu_dict
```

```python
for n in pivot.index:
    if pivot['광역시도'][n][-3:] not in ['광역시', '특별시', '자치시']:
        if pivot['시도'][n][:-1]=='고성' and pivot['광역시도'][n]=='강원도':
            si_name[n] = '고성(강원)'
        elif pivot['시도'][n][:-1]=='고성' and pivot['광역시도'][n]=='경상남도':
            si_name[n]='고성(경남)'
        else:
            si_name[n] = pivot['시도'][n][:-1]
            
        for keys, values in tmp_gu_dict.items():
            if pivot['시도'][n] in values:
                if len(pivot['시도'][n])==2:
                    si_name[n] = keys + ' ' + pivot['시도'][n]
                elif pivot['시도'][n] in ['마산합포구', '마산회원구']:
                    si_name[n] = keys + ' ' + pivot['시도'][n][2:-1]
                else:
                    si_name[n] =keys + ' ' + pivot['시도'][n][:-1]
                
    elif pivot['광역시도'][n] == '세종특별자치시':
        si_name[n] = '세종'
        
    else:
        if len(pivot['시도'][n])==2:
            si_name[n] = pivot['광역시도'][n][:2] + ' ' + pivot['시도'][n]
        else:
            si_name[n] = pivot['광역시도'][n][:2] + ' ' + pivot['시도'][n][:-1]

si_name
pivot['ID'] = si_name
pivot.head()
```

```python
import folium
import json
import warnings

warnings.simplefilter(action='ignore', category = FutureWarning)

pivot_folium = pivot.set_index('ID')
pivot_folium
```

```python
geo_path = 'C:/Users/tjdwo/pydata/05. skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding = 'utf-8'))

map = folium.Map(location=[36.2002, 127.054], zoom_start=7)
map.choropleth(geo_data = geo_str,
              data = pivot_folium['인구수합계'],
              columns = [pivot_folium.index, pivot_folium['인구수합계']],
              fill_color = 'YlGnBu' , #PuRd, YlGnBu
              key_on = 'feature.id')
map
```

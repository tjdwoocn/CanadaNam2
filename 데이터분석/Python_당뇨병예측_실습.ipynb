```python
from html.parser import HTMLParser
import urllib.request

TOPIC = "Decision tree"
url = "https://en.wikipedia.org/wiki/%s" % TOPIC
class LinkCountingParser(HTMLParser):
    in_paragraph=False
    link_count=0
    def handle_starttag(self,tag,attrs):
        if tag=='p':
            self.in_paragraph = True
        elif tag =='a' and self.in_paragraph:
            self.link_count += 1
    def handle_endtag(self, tag):
        if tag=='p':
            self.in_paragraph = False
```

```python
html= urllib.request.urlopen(url).read()
print(html)
html = html.decode('utf-8')
parser = LinkCountingParser()
parser.feed(html)
print("there were", parser.link_count, "links in the article")
```

# 당뇨병 예측하기 (회귀분석)

```python
import sklearn.datasets
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import normalize
from sklearn.metrics import r2_score
```

```python
#1. data load
diabetes = sklearn.datasets.load_diabetes()
#2. normalize
X, Y = normalize(diabetes['data']),diabetes['target']

#3. train-set, test- set
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=.7)

#4. regression model
linear = LinearRegression()
linear.fit(X_train, Y_train)

#5. Predict
preds_linear = linear.predict(X_test)

#6-1 get correlation
corr_linear= round(pd.Series(preds_linear).corr(pd.Series(Y_test)),3)
print(corr_linear)

#6-2 get R2 (validation)
rsquared_linear = r2_score(Y_test, preds_linear)
print(rsquared_linear)
```

```python
#1. data load
diabetes = sklearn.datasets.load_diabetes()
print(diabetes)
print(diabetes['data'][0:10])
```

```python
df = pd.DataFrame(diabetes['data'])
df.head()
```

```python
# AGE, SEX, BMI, BP, S1, S2, S3, S4, S5, Y
df.rename(columns={df.columns[0]: 'AGE'}, inplace=True)
df.rename(columns={df.columns[1]: 'SEX'}, inplace=True)
df.rename(columns={df.columns[2]: 'BMI'}, inplace=True)
df.rename(columns={df.columns[3]: 'BP'}, inplace=True)
df.rename(columns={df.columns[4]: 'S1'}, inplace=True)
df.rename(columns={df.columns[5]: 'S2'}, inplace=True)
df.rename(columns={df.columns[6]: 'S3'}, inplace=True)
df.rename(columns={df.columns[7]: 'S4'}, inplace=True)
df.rename(columns={df.columns[8]: 'S5'}, inplace=True)
df.rename(columns={df.columns[9]: 'Y'}, inplace=True)

df.head()
```

```python
#2 normalize
X,Y = normalize(diabetes['data']), diabetes['target']
print(X[0:10])
print(Y[0:10])
```

```python
#3 train-set, test-set
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=.7)
print(len(X_train))
print(len(X_test))
print(len(Y_train))
print(len(Y_test))
```

```python
#4 regression model

linear=LinearRegression()
print(linear)
linear.fit(X_train, Y_train)

#5 predict

preds_linear= linear.predict(X_test)
print(preds_linear[0:10])
```

```python
# 6-1 get correlation
corr_linear = round(pd.Series(preds_linear).corr(pd.Series(Y_test)),3)
print(corr_linear)

# 6-2 get R2
rsquared_linear = r2_score(Y_test, preds_linear)
print(rsquared_linear)
```

```python
print('Linear Coefficients')
print(linear.coef_)
print(linear.intercept_)
plt.scatter(preds_linear, Y_test)
plt.title('Lin. Reg. Corr= %f Rsq = %f' % (corr_linear, rsquared_linear))
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.plot(Y_test, Y_test, 'k--')

# predicted = actual line, for comparison
plt.show()
```
